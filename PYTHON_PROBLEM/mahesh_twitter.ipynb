# Authenticating twitter API

```python
  
    #After getting the token  we have to create the instance of oAuthHandler ,Into this we pass our consumer token 
    # and secret which was given to us .    
auth = OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_key,access_secret)

api=tweepy.API(auth)
```

```python

def process_media(entities):
  #accepts a tweet entities object,
  #returns the number of images in the tweet.
  
  ret = 0
  
  if entities.get('media'):
    ret += len(entities.get('media'))
  
  return ret if ret else None
```

```python
#number of tweets n
n=300

tlst=[]    
#here tweepy.cursor is a inbuilt method in which i have passed api.search which will search the particular word passed ,like here
# i have passed # car, so it will give all searches where that particular word had ocuur and 
# rpp(return per page)-The number of tweets to return per page, up to a max of 100

for tweet in tweepy.Cursor(api.search,q="#car",rpp=100).items(n):
                           
                           tweetjson=tweet._json                 #here creating json object 
                           jfile=open('jtweet.json','a')          # a file name jtweet.json is created and a is append mode.
                           json.dump(tweetjson,jfile,sort_keys=True,indent=4)   #dump() is basically used for writing data
                           #If sort_keys is True (default False), then the output of dictionaries will be sorted by key
                          
                           print("\n")
                           print(tweetjson['created_at']+"    "+tweetjson['text']+"    "+str(tweetjson['retweet_count'])+"    "+str(tweetjson['favorite_count'])+"    "+str(tweetjson['entities']))                                 
                           print("\n")
                        #Now these are the attributes for retrieving the data which will be appended in the tlst dictionary list.
                           
                           time=str(tweetjson['created_at'])
                           text=str(tweetjson['text'])
                           retweets=str(tweetjson['retweet_count'])
                           likes=str(tweetjson['user']['favourites_count'])
                           entities=str(tweetjson['entities'])
                        
                           tlst.append([time,text,retweets,likes,entities])
jfile.close()
```

```python
#dataframe created and pases the list to it.
df=pd.DataFrame(tlst,columns=['Time','Tweets','Retweets','Likes','entities']).astype(str)
print(df)  
df.to_csv('finaldata.csv')   #change to csv file
```
